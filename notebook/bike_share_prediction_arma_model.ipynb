### Import Libraries

```python
hourly_data = pd.date_range('2011-01-01', periods = new_total_instances, freq = 'H')
hourly_data = pd.DataFrame(hourly_data)
hourly_data.columns = ['datetime']
hourly_data
```

```python
# Create new columns for year, month, day and hour for the dataframe
df['year'] = df['dteday'].dt.year
df['month'] = df['dteday'].dt.month
df['day'] = df['dteday'].dt.day
df['hour'] = df['hr']
df.tail(10)
```

```python
# Concat dataframe into 'datetime' and remove unnecesarry columns
df['datetime'] = pd.to_datetime(df[['year', 'month', 'day','hour']])
df = df.drop(['instant' ,'dteday', 'yr','mnth','hr','year', 'month', 'day','hour'], axis = 1)
df
```

```python
# Merge the missing timestamp onto dataframe
df_new = pd.merge(df,hourly_data,on='datetime',how='outer')
df_new
```

### Model Preparation

```python
df_model = pd.Series(df_clean.cnt.values, index=df_clean.index)
df_model
```

```python
def print_adf_result(adf_result):
    df_results = pd.Series(adf_result[0:4], index=['ADF Test Statistic','P-Value','# Lags Used','# Observations Used'])
    for key, value in adf_result[4].items():
        df_results['Critical Value (%s)'% key] = value
    print('Augmented Dickey-Fuller Test Results:')
    print(df_results)
result = adfuller(df_model)
print_adf_result(result)
```

It is stationary when the P value is less than 0.05

#### Split Train and Test

```python
split_ratio = round(df_model.shape[0]*0.8)
train_use = df_model.iloc[:split_ratio]
test_use = df_model.iloc[split_ratio:]
test_use.isnull().values.any()
```

### Model Forecast

```python
# Build Model
arma_model = ARMA(df_model, order=(6, 5)).fit()

# Forecast
fc, se, conf = arma_model.forecast(3509, alpha=0.05)  # 95% conf

# Make as pandas series
fc_series = pd.Series(fc, index=test_use.index)
lower_series = pd.Series(conf[:, 0], index=test_use.index)
upper_series = pd.Series(conf[:, 1], index=test_use.index)

# Plot
plt.figure(figsize=(12,5), dpi=100)
plt.plot(train_use, label='training')
plt.plot(test_use, label='actual')
plt.plot(fc_series, label='forecast')
plt.fill_between(lower_series.index, lower_series, upper_series, 
                 color='k', alpha=.15)
plt.title('Forecast vs Actuals')
plt.legend(loc='upper left', fontsize=8)
plt.show()
```

### Model Evaluation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
print(f'Mean Absolute Error = {mean_absolute_error(test_use,fc)}')
print(f'Mean Squared Error = {mean_squared_error(test_use,fc)}')
print(f'Root Mean Squared Error = {math.sqrt(mean_squared_error(test_use,fc))}')
```
