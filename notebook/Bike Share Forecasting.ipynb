However to be sure, we can also create a boolean mask of whether that particular datapoint is a outlier or not. To do this, we need to:

1. Compute the interquartile range (IQR)
2. Determine whether the particular data point is < 1.5 * IQR or > 1.5 * IQR

```python
df_quartile = df_interpolate.quantile([.25, .75])
df_quartile
```

```python
df_quartile.iloc[1] - df_quartile.iloc[0]
```

```python
def compute_outlier_threshold(data, lower_threshold=True):
    q25, q75 = np.quantile(data, .25), np.quantile(data, .75)
    iqr = q75 - q25
    lower, upper = q25 - 1.5*iqr, q75 + 1.5*iqr
    if lower_threshold:
        return lower
    else:
        return upper
```

```python
df_outlier = pd.DataFrame()
```

```python
for col in df_interpolate.drop('datetime', axis=1).columns:
    df_outlier[col] = df_interpolate[col].apply(lambda x: True if ((x < compute_outlier_threshold(df_interpolate[col])) or (x > compute_outlier_threshold(df_interpolate[col], False))) else False)
```

```python
df_outlier
```

```python
df_outlier.sum()/len(df_outlier)*100
```

```python
# Outlier Detection using Inter Quartile Range
def out_iqr(s, k=1.5, return_thresholds=False):
    """
    Return a boolean mask of outliers for a series
    using interquartile range, works column-wise.
    param k:
        some cutoff to multiply by the iqr
    :type k: ``float``
    param return_thresholds:
        True returns the lower and upper bounds, good for plotting.
        False returns the masked array 
    :type return_thresholds: ``bool``
    """
    # calculate interquartile range
    q25, q75 = np.percentile(s, 25), np.percentile(s, 75)
    iqr = q75 - q25
    # calculate the outlier cutoff
    cut_off = iqr * k
    lower, upper = q25 - cut_off, q75 + cut_off
    if return_thresholds:
        return lower, upper
    else: # identify outliers
        return [True if x < lower or x > upper else False for x in s]
    
    
# For comparison, make one array each at varying values of k.
df4 = df_interpolate.drop(columns=['datetime'],axis=1)
iqr1 = df4.apply(out_iqr, k=1.5)
iqr1.head(10)
```

### Moving Average Method

Under the moving average forecasting methods, we have:
- Single exponential smoothing (SES) aka Brown's linear
- Double exponential moving average
    - Holt's linear trend
    - Holt's damped trend
- Triple exponential moving average aka Holts-Winter method

However, all are not created equally, as some methods suit a particular type of data. For example:
- Single exponential smoothing (SES) aka Brown's linear: suitable for series with no trend and seasonality
- Double exponential moving average: suitable for series with trend but no seasonality
- Triple exponential moving average aka Holts-Winter method: suitable for series with trend and seasonality.

Therefore, we are only going to perform Holts-Winter method.

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
test_num = 200

df_holt_train = df_timeseries[['datetime', 'cnt']].iloc[:len(df_timeseries) - test_num]
df_holt_train = df_holt_train.set_index('datetime')
df_holt_train
```

```python
df_holt_test = df_timeseries[['datetime', 'cnt']].iloc[len(df_timeseries) - test_num:]
df_holt_test = df_holt_test.set_index('datetime')
df_holt_test
```

Holt-Winter - Trend: Additive, Seasonal: Additive

```python
holt_add_add = ExponentialSmoothing(pd.Series(df_holt_train['cnt'], df_holt_train.index), trend='add', seasonal='add', seasonal_periods=24, initialization_method='estimated').fit()
holt_add_add_fcast = holt_add_add.forecast(test_num).rename('trend_add_seasonal_add')
holt_add_add_fcast
```

```python
holt_add_add.summary()
```

```python
plt.figure(figsize=(20,10))
plt.plot(df_holt_test, color='black', marker='o', linestyle='dashed')
plt.plot(holt_add_add_fcast, color='blue', marker='o', linestyle='dashed')
plt.legend(['test', holt_add_add_fcast.name])
```

```python
holt_add_mul = ExponentialSmoothing(pd.Series(df_holt_train['cnt'], df_holt_train.index), trend='add', seasonal='mul', 
                                    seasonal_periods=24, initialization_method='estimated').fit()
holt_add_mul_fcast = holt_add_mul.forecast(test_num).rename('trend_add_seasonal_mul')
holt_add_mul_fcast
```

```python
holt_add_mul.summary()
```

```python
plt.figure(figsize=(20,10))
plt.plot(df_holt_test, color='black', marker='o', linestyle='dashed')
plt.plot(holt_add_mul_fcast, color='blue', marker='o', linestyle='dashed')
plt.legend(['test', holt_add_mul_fcast.name])
```

```python
holt_mul_add = ExponentialSmoothing(pd.Series(df_holt_train['cnt'], df_holt_train.index), trend='mul', seasonal='add', damped_trend=True,
                                    seasonal_periods=24, initialization_method='estimated').fit()
holt_mul_add_fcast = holt_mul_add.forecast(test_num).rename('trend_mul_seasonal_add')
holt_mul_add_fcast
```

```python
holt_mul_add.summary()
```

```python
plt.figure(figsize=(20,10))
plt.plot(df_holt_test, color='black', marker='o', linestyle='dashed')
plt.plot(holt_mul_add_fcast, color='blue', marker='o', linestyle='dashed')
plt.legend(['test', holt_mul_add_fcast.name])
plt.show()
```

```python
holt_mul_mul = ExponentialSmoothing(pd.Series(df_holt_train['cnt'], df_holt_train.index), trend='mul', seasonal='mul', damped_trend=True,
                                    seasonal_periods=24, initialization_method='estimated').fit()
holt_mul_mul_fcast = holt_mul_mul.forecast(test_num).rename('trend_mul_seasonal_mul')
holt_mul_mul_fcast
```

```python
holt_mul_mul.summary()
```

```python
plt.figure(figsize=(20,10))
plt.plot(df_holt_test, color='black', marker='o', linestyle='dashed')
plt.plot(holt_mul_mul_fcast, color='blue', marker='o', linestyle='dashed')
plt.legend(['test', holt_mul_mul_fcast.name])
plt.show()
```

```python
print("AIC, BIC and AICC Values of Each Config")
print("Additive trend, additive season: ", holt_add_add.aic, ", ", holt_add_add.bic, ", ", holt_add_add.aicc)
print("Additive trend, multiplicative season: ", holt_add_mul.aic, ", ", holt_add_mul.bic, ", ", holt_add_mul.aicc)
print("Multiplicative trend, additive season: ", holt_mul_add.aic, ", ", holt_mul_add.bic, ", ", holt_mul_add.aicc)
print("Multiplicative trend, multiplicative season: ", holt_mul_mul.aic, ", ", holt_mul_mul.bic, ", ", holt_mul_mul.aicc)
```

Holt-Winter Exponential Smoothing with multiplicative trend and season is the best model, with lowest AIC, BIC and AICC.

### AR, MA, ARIMA and SARIMA

```python
from statsmodels.tsa.stattools import adfuller
```

```python
def adf_test(timeseries):
    adf = adfuller(timeseries)
    print("ADF Test Statistic: ", adf[0])
    print("p value: ", adf[1])
    print("Lags: ", adf[2])
    print("Number of observations: ", adf[3])
    print("Critical value (1%): ", adf[4]['1%'])
    print("Critical value (5%): ", adf[4]['5%'])
    print("Critical value (10%): ", adf[4]['10%'])

adf_test(df_timeseries['cnt'])
```

Surprisingly, the original time series data is stationary, because:
- Test statistics is less than critical values
- P value is less than predetermined alpha=0.05

Hence, we reject the null hypothesis.
